4. Cleaning and Preprocessing

A. Cleaning

```python
## Inspecting for null values

data.isna().sum()
```

```python
## Inspecting Duplicate Values

data.duplicated().sum()
```

```python
duplicates = data.duplicated( subset = ['Location', 'Date_Time' ]).sum()
```

```python
print(duplicates)
```

```python
duplicates = data[data.duplicated( subset = ['Location', 'Date_Time' ], keep = False)]
```

```python
duplicates.reset_index(drop = True)
```

```python
duplicates.head(50)
```

```python
##Dropping Duplicates

cln_data = data.drop_duplicates( subset = ['Location', 'Date_Time'])
```

```python
cln_data.head(10)
```

```python
cln_data
```

```python
cln_data1 = cln_data.reset_index(drop = True)
```

```python
cln_data1
```

```python
cln_data1['Date_Time'].dtype
```

```python
## Converting to datetime datatype

cln_data1['Date_Time'] = pd.to_datetime(cln_data1['Date_Time'], errors='coerce')
```

```python
cln_data1['Date_Time'].dtype
```

```python
## Splitting dates and time from the column

cln_data1['Date'] = cln_data1['Date_Time'].dt.date
cln_data1['Time'] = cln_data1['Date_Time'].dt.time
```

```python
cln_data1.head(10)
```

```python
## Dropping the Date_Time column

cln_data1.drop(columns = ['Date_Time'], inplace = True)
```

```python
cln_data1.head(10)
```

```python
## Resetting column index

columns_to_move = ['Date', 'Time']
columns = [col for col in cln_data1.columns if col not in columns_to_move]
columns[1:1] = columns_to_move
cln_data1 = cln_data1[columns]
```
